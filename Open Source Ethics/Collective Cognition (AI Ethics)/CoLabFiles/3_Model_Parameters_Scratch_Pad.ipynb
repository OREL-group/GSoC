### Case Study \#2 GitCompare

[GitCompare](https://gitcompare.com/) uses the following to generate a score for a given GitHub repository:

* **Activity** - Represents a repository general development activity, influenced by the amounts of commits, forks, and time from last commit.
* **Community** - Represents a repository community strength, influenced by the amounts of stars, watchers, forks, and contributors.
* **Engagement** - Represents how people engage with a repository, influenced by the amounts of stars, watchers, forks, contributors, issues, and PRs.
* **Maintenance** - Represents well maintained a repository is, influenced by the amounts of contributors, issues, and PRs(Open and closed).
* **Adoption** - In case we have "downloads" information this represent a relative adoption score.

This is nice, it's difficult to find a nice explanation of their algorithm, or a list of the weights as in SourceCred, but breaking it down like this is perhaps a good way of examining the end result of community engagement rather than the engagement itself.

### Trying to find the parameters of Collective Cognition

The Santa Fe workshop referenced above seems to point to Multi-Agent Systems for Cognition as being a mathematically productive place to start. This is also nice with our interest in Dec-POMDPs and Active Inference/Free Energy Principle.

The workshop proposes Collective Cognition as "an interaction among three elements-the individual abilities of the agents, their shared knowledge, and their communication structure" or put in list form:
* individual abilities
* shared knowledge
* communication structure

Which in Open Source Communities maps (one possibility of many) to:
* individual abilities -- code commits
* shared knowledge -- shared community goals/intentions
* communication structure -- roles in organization

They then diagnose the "forward problem" of advancing the shared knowledge of the collective, with a few ways of stating the issue:
* "we wish, for instance, to know just how analogous to organic evolution this knowledge evolution really is"
* "the less well-understood inverse problem: starting with a desired global behavior for a collective, design the goals of constituent agents so as to produce that behavior."

## Still searching for Collective Cognition
### a functional description - no more intentionality discussion (for now)

Intention as described by Tuomela and Miller (and others) is only part of what we are looking at here. It is (almost) a given that in our Open Source Software Development Communities all involved intend to create or develop software.

We could use Tuomela and Miller's formulation of "we-intention" to determine who is or is not "part of" the collective cognition and/or intelligence which we are looking at. To summarize again in list form the criteria for joint intention or we-intention of an agent:
* each agent has the we-intention to perform X act
* there is a mutual belief that the agent intends to X, and that all group members will X
* the agent believes that all group members believe that X is possible

These could be used to sort out "bad actors" - though without access to their internal state (and this seems to be Searle's point about reduction), we cannot use this argument functionally as Tuomela indicates. From our perspective as constructors of a computational model we have no ability to access the intentions of our agents, only to model their behaviors.

Before letting the intentionality angle go for a while, I do want to note its strong connection to ideas of an Ethical System and correspondences to the enactive approach. It seemed probable/possible that such a functional approach from a philosophical angle could be fruitful when applied, and it may still be, but for now let's dig into the Active Inference approach instead.


### Looking in Philosophy

Also of interest is the phrase ["Collective Intentionality" as opposed to "Collective Cognition" - a la the Stanford Encyclopedia of Philosophy](https://stanford.library.sydney.edu.au/entries/collective-intentionality/) - which seems to be related and/or the same as the phrase "Collective Cognition" which we use here.

Of the many philosophers referenced here, some things stick out:
* Scheler's "joint intentionality" and that it is NOT a combination of individual intentionality with a sructure of reciprocal awareness. Instead, for Scheler, it is that "many minds are in a numerically indentical state" - an "irreducible collective" - such extremes are perhaps more of a disraction than we need for a computational model, but food for thought nonetheless
* Sellar's "we-intentions" are perhaps more practical for our purposes: "We-intentions are attitudes, but at the same time, they are not merely private, but involve a shared point of view from which the participants may critically assess each other’s contributions."
* Also appropriately enough for Sellars, "only the we-intentions that are universally shared, and no parochial group attitudes, constitute the moral point of view."

There are some interesting parallels in the Intentionality argument with DePraz, et al in "On Becoming Aware" as well, for instance the appreciation of the necessity of attention to the act (which is collectively intended) and the mode (the collective or "we-mode"). This is followed through after beginning with Sellars' "we-intentions" by Tuomela and Miller but phrased as "a special kind of intention had by an individual agent."

Of all people, John Searle enters the discussion at this point, taking issue with Tuomela and Miller's attempt to "reduce collective intentions to individual intentions plus beliefs." Tuomela in particular rejects this analysis by saying it "is rather meant to elucidate the irreducible notion of we-intention in a functionally informative way" - which brings us full circle back to Scheler's notion of the irreducible, but with this nod to the we-intention in the individual we break the irreducibility.

Tuomela's formulation of "joint intention":


    (JI) Agents A1, …, Ai, …, Am have the joint intention to perform a joint action X if and only if

        these agents have the we-intention (or are disposed to form the we-intention) to perform X; and
        there is a mutual belief among them to the effect that (a).

    In the case of joint intention the conatively used ‘We will do X’ is true of each participant Ai. (Tuomela 2005: 342)

Appropriately they use the $A_i$ term, which I will note here as a joke.

Regardless, summarizing this Encyclopedia article is interesting, but not so helpful to our attempt to find parameters of Collective Cognition.

## Active Inference and Collective Cognition

*All page numbers reference ["Active Inference"](https://mitpress.mit.edu/books/active-inference) by Parr et al*

Taking as a starting point (another one) the "recipe" given for designing an Active Inference Model to see if that's even an appropriate model for our purposes, let's look at the four foundational steps (p 106) in Friston, et al:
* Which system are we modeling?
* What is the most appropriate form for the generative model?
* How to set up the generative model?
* How to set up the generative process?

### Which system are we modeling?
* Identification of boundaries
    * This is meant to set up the internal/external divide for our system, for an Open Source Community this could be the community itself and the "world-at-large" but it could also be the information/data we have for that community, the data we have access to, and the "unknowns" which exist beyond it.
    * The big question for the boundaries here are related to whether we see the entire community as an agent, or whether we see each member as an agent forming the community! With the idea of collective cognition this gets a bit more complex.
    * At the moment let's assume that our collective cognition is made up of agents, and that these agents are the members of the Open Source Community
        * Just a note about this assumption, which seems increasingly appropriate as we dive in, which is that if we see the Open Source community/project as "world" then we can talk about the internal models of each agent of that world, which gels nicely with our philosophical research and Active Inference's framework!
    * So the Markov Blanket we will use here would be the separator between agents in the community and the community-at-large
* What counts as an Active Inference agent
    * In our model our agents are the individuals involved in the community
    * The quantitative aspects would be an agent's commits/activity, behavior traces, location, times, burstiness, forks, issue resolution, and probably much more
* What counts as the External environment
    * The agent's world is the community itself, the project's parameters, goals, riskiness, growth, demographics, ethics, leadership, and much more
* What is the interface between them
    * Here it would be nice to fold in an embodied approach, which is mentioned briefly in Active Inference p 109
    * Here is where our discussion above of the Markov Blanket could be fleshed out - to include flesh/sensori-motor considerations
    * However it occurs to me at this point that although an embodied approach would be great to include, the data we're utilizing is really from the DISembodied data which each agent produces, not their embodied/sensori-motor data
    * Will have to develop the interface more to clearly delineate what is being inferred (the state of the community according to its shared goals) and what is doing the inferring (each agent while they participate or do not).

### What is the most appropriate form for the generative model?
* Are the inferences Categorical, Discrete, or a Hybrid?
    * Here a hybrid approach is the most tempting, given the wide scope of our parameters.
* What is the most appropriate hierarchical depth?
    * For our model the hierarchical depth will be quite complex, involving transit between the various types of data and considerations
* Do we include temporal depth and the ability to predict action-contingent observations to support planning?
    * Temporal depth is crucial to our understanding, as is planning, so yes!

### How to set up the generative model?
* Most appropriate variables and priors?
    * Using CHAOSS's schema, we would use the following:
        * Metrics:
            * Contributions
            * Time
            * People
            * Place
        * Diversity, Equity and Inclusion:
            * Event Diversity
            * Governance
            * Leadership
            * Project and Communities
        * Evolution
            * Code Development Activity
            * Code Development Efficiency
            * Code Development Process Quality
            * Issue Resolution
            * Community Growth
        * Risk
            * Business Risk
            * Code Quality
            * Dependency Risk Assessment
            * Licensing
            * Security
        * Value
            * Academic
            * Communal
            * Individual
            * Organizational
    * However let's cherry-pick from this for our variables and priors, as it is not feasible to use all of these for our current project, it would be too bloated and non-specific/non-quantifiable/non-computational:
        * Development Variables:
            * Activity
            * Efficiency
            * Process Quality
            * Issue Resolution
            * Security
        * Social Variables:
            * Contributions
            * Time

            * Place
        * Ethical Priors:
            * Diversity/Variability

    * A note about this structure, which includes Variables with two aspects, Developmental and Social - the variables are demarcated primarily to emphasize their source and relationship to an Open Source Software Community. Also the Priors seem like a great place to consider our Ethically Regulated System's unambiguous and transparent goals, the Good Regulator Theorem is even mentioned in Active Inference on p 117 as being especially appropriate for the Priors!
    * A consideration here is how to implement the priors here computationally. We are imagining our community and its agents as an organism with affordances and perceptions. A picture of this might be nice.
* Which parts are fixed and which must be learned?
    * The Priors and Community Goals are relatively fixed but nearly everything here is learned/evolves temporally.

### How to set up the generative process?
* What are the elements of the process?
    * As stated in Active Inference, the process is typically nearly identical to or closely related to the model outlined above. However for our model, the process is that of Collective Cognition, wherein all agents work together to form a Cognizing Community.
* How do they differ from the model?
    * In this respect, that of Collective Cognition, the elements of the process have a "macro" focus in a way that the agents do not. For each agent, there is an internal and external set of variables and priors, beliefs and percepts, actions and defining tendencies. For the process as a whole, the conglomeration of the agents forms a single collective agent for whom the external is the world-at-large!

## The world

The tutorial uses a 'simple grid-world' for its agent to exist in. For us, our world is considerably more complex and subtle - not simply renderable in 2 dimensions.

### Parameters of the world

From our recipe above, we have identified conceptually what the world is, namely the world of an Open Source Software Development Community. But what, precisely are its parameters?

If we can view an Open Source Project as the 'world' - then its aspects are its parameters:
*    Working, completed, reviewed code
*    Repository maintenance tasks (dependencies, security)
*    Code which is in the Pull Request state

This is one possible formulation, by which the agents "move" in this world by their work with each other on the creation and modification of the contents of the repository.

Given these three possible 'areas' of the world, it's possible to conceive of a repository as a three-dimensional graph with the axes representing code which is part of the repo, code which performs maintenance, and code which requires review or is being modified.

In this model, the agents move by working on the codebase which they all share. Effectively working as one cognizing system to realize a shared goal.


### Defining Sustainability

Following AlMarzouq et al in their ["Taxing the development structure of open source communities: An information processing view"](http://dx.doi.org/10.1016/j.dss.2015.09.004) when they note: "when FLOSS communities do not develop highly modular codebase,
we expected that centralized committal structures are more apt to lead to higher performance" here our focus is on encouraging sustainability. AlMarzouq et al find a tension between Raymond and Brooks and ultimately state:
> More importantly, we consolidate the views forwarded by Brooks [12] and Raymond [81] about software development. While Raymond [81] posited that task independence is a key assumption for the bazaar
model to thrive, we provide empirical evidence to support this idea, and identify other factors that come into play. In fact, under certain conditions, we demonstrate that FLOSS communities are similar to
traditional software development teams and that Brooks' law [12], as one of the most important classical theories on software project management, still holds true in the FLOSS context.

[12] F. Brooks, The mythical man-month, Proceedings of the international conference on reliable software, vol. 10, ACM Press, 1975.

[81] E. Raymond, The cathedral and the bazaar: musings on Linux and open source by an accidental revolutionary, revised edition O'Reilly, Cambridge, MA, 2001.

So in our search for how to define sustrainability, we can say that part of it depends on the centralization of the community and the modularity of the code, thanks to AlMarzouq et al.
